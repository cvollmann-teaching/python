# Wissenschaftliches Rechnen mit Python: The SciPy Stack

Scientific computing in Python is done using the **SciPy Stack** (https://scipy.org/). This term is used for a collection of packages, developed for scientific computing. From this collection we will mainly use the following three packages:

- **Numpy** (the basis)
     - provides the flexible data type `numpy.ndarray` (e.g., for matrices and vectors)
     - contains a huge amount of tools to perform all sorts of array manipulation (hardware aware)



- **SciPy** (the core)
    - builds upon data objects from numpy
    - contains a huge amount of numerical methods (solving linear systems, optimization, integration, interpolation,...)
    - **Do not reinvent the wheel!** In practice, do not use self-implemented routines, when they are available in SciPy (under the hood SciPy uses well etablished FORTRAN routines from LAPACK, BLAS,...)
    

- **Matplotlib** (the visualizer)
    - allows to visualize results with high quality grafics (plots in 2d and 3d, images and videos,...)  

Remarks:
- **Tutorial:** the official one can be found here: https://scipy-lectures.org/. The first chapter in this tutorial contains already enough to do basic math related programming. 


- **Installation:** typically not needed! They are usually contained in the system-side Python installation or come along with the respective distribution (e.g., anaconda). However, for python venv typically manual installation needed


- **Import convention:** a program typically starts with

## Appetizer

````python
# Oftmals der Start eines Numerik-Skripts
import numpy as np
import scipy as sp 
import matplotlib as mpl
import matplotlib.pyplot as plt
````

````python
def f(x):
    return x**2
X = np.linspace(-1,1,100) 
#print(X)
Y = f(X)
#print(Y)
plt.plot(X,Y)
plt.show()
````

### Optimierung 
$\min_x f(x)$

````python
from scipy import optimize
x_min  = optimize.fmin(f, 0.1)
print(x_min)
````

### Integration
$\int_a^b f(x)dx$

````python
from scipy import integrate
integral = integrate.quad(f,-1, 1)
print(integral)
````

### Interpolation

````python
from scipy.interpolate import interp1d
f_interp = interp1d(X,Y)
X2 = np.linspace(-1, 1, 10)
plt.plot(X, Y, '-', X2, f_interp(X2), 'r--', X2, f_interp(X2), 'rD')
plt.show()
````

### Spline Interpolation

````python
from scipy.interpolate import splprep, splev
t = np.arange(0, 1.1, .1)
x, y = np.sin(2*np.pi*t), np.cos(2*np.pi*t)
tck, u = splprep([x, y], s=0)
unew = np.arange(0, 1.01, 0.01)
out = splev(unew, tck)
plt.plot(x, y, 'x', np.sin(2*np.pi*unew), np.cos(2*np.pi*unew), x, y, 'b')
plt.show()
````

### Lineare Gleichungssysteme

Löse $Ax = b$

````python
dim = 10
A = np.eye(dim,dim, dtype =int)
A[tuple(i+1 for i in range(dim-1)), tuple(i for i in range(dim-1))] = -np.ones(dim-1)
A = (A + A.T)
print(A)
b = np.ones(dim)
x = sp.linalg.solve(A,b)
plt.plot(np.linspace(0,1,dim),1./dim**2* x)
````

### LR Zerlegung (Gaußsches Eliminationsverfahren)

Für jede Matrix $A \in \mathbb{R}^{m\times n}$ finden wir eine Zerlegung
$$P A = LU $$
wobei
- $P$ Permutation matrix
- $L$ linke untere Dreiecksmatrix
- $U$ Zeilenstufenform

````python
P, L , U = sp.linalg.lu(A)              # Zerlegung
print(np.allclose(A.dot(P), L.dot(U)))  # Test ob PA = LU
#A@P
````

### Toeplitz solver

````python
plt.imshow(A)
````

````python
x = sp.linalg.solve_toeplitz(A[:,0],b)
plt.plot(np.linspace(0,1,dim), 1./dim**2* x)
````

### Sparse Formate

````python
A_sparse = sp.sparse.csr_matrix(A)
print(A_sparse.indices)
print(A_sparse.indptr)
print(A_sparse.data)
x = sp.sparse.linalg.spsolve(A_sparse, b)
plt.plot(np.linspace(0,1,dim),1./dim**2* x)
````

### Kleinste Quadrate Lösung

$$ x = \arg \min_x \|Ax-b \|_2^2$$

 - Notwendige Optimalitätsbedingung: Normalengleichung
     $$A^T A x = A^Ty $$

````python
x = np.linalg.lstsq(A,b, rcond=None)[0]
plt.plot(np.linspace(0,1,dim),1./dim**2* x)
````

### Ausgleichsrechnung mit kleinste Fehlerquadrate (curve fitting)

 - Gegeben: Input/Output - Daten $$(x_i, y_i) ,~~~~  i=1, \ldots, n$$
 - Lineares Model: $$m_{a,b}(x) = ax + b \approx y$$
 - Ziel: Finde Parameter $a$ und $b$, sodass Residuum minimal (in der Euklischen Norm), d.h.,
     $$(a,b) = \arg \min_{(a,b)} \sum_{i=1}^n (m_{a,b}(x_i)-y_i)^2 = \arg \min_{(a,b)} \|M\left(\begin{array}{rr}a \\b\end{array}\right) - y \|_2^2  $$
 - Notwendige Optimalitätsbedingung: Normalengleichung
     $$M^T M x = M^Ty $$


````python
import random
n = 60
x_i = np.linspace(-3,3,n)
y_i = 1.5 * x_i - 1 + np.array([random.uniform(-1, 1) for i in range(n)])
M = x_i[:,np.newaxis]**[0,1]
[b,a], residuals, rank, svals = np.linalg.lstsq(M, y_i, rcond=None) # solve the lstsq problem with scipy
t = np.linspace(-3,3,100)
plt.plot(t, a*t + b , x_i, y_i, "o"), plt.show()
````

### Eigenzerlegung symmetrischer Matrizen

Für jede *symmetrische* Matrix $A \in \mathbb{R}^{n \times n}$ existieren $\Lambda \in \mathbb{R}^{n \times n}$ und $V\in \mathbb{R}^{n \times n}$ mit
 $$ A= V\Lambda V^T,$$ wobei
 $\Lambda$ Diagonalmatrix (enthält Eigenwerte) und $V$ orthogonal (Spalten = Eigenvektoren).

````python
lamb, V = np.linalg.eigh(A) # Eigenzerlegung berechnen mit NumPy
plt.plot(lamb)
print(np.allclose(A, V.dot(np.diag(lamb).dot(V.T)))) # Teste ob A = V Lambda V^T
np.allclose(np.linalg.det(A),np.prod(lamb)) # Teste ob det(A) = \prod_i eigvals_i
print(min(lamb))
````
