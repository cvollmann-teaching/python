# matplotlib

Tutorial: https://scipy-lectures.org/intro/matplotlib/index.html#matplotlib

````python
import matplotlib.pyplot as plt
````

`plt.plot()`

doc = https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

````python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
plt.plot(X, np.cos(X), X, np.sin(X))
plt.show()
````

````python

# Create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8, 6))

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

# create equidistant grid
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# Plot cosine with a blue continuous line of width 1 (pixels)
plt.plot(X, np.cos(X), color="blue", linewidth=1.0, linestyle="-", label="cosine")

# Plot sine with a green continuous line of width 1 (pixels)
plt.plot(X, np.sin(X), color="green", linewidth=1.0, linestyle="-", label="sine")

# -- X ACHSE --
# Set x limits
plt.xlim(-4.0, 4.0)
# Set x ticks
#plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# -- Y ACHSE --
# Set y limits
plt.ylim(-1.0, 1.0)
# Set y ticks
#plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

# Spines = lines connecting the axis tick
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('red')
ax.spines['top'].set_color('yellow')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0.))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',-0.5))
ax.spines['right'].set_color('red')

# legend
plt.legend(loc='upper left')

# annonate

# Show result on screen
plt.show()
````

## figure, subplots, axes und ticks

![image-axes](https://www.math.uni-trier.de/~vollmann/prog/img/axes.png)https://www.math.uni-trier.de/~vollmann/prog/img/axes.png

- `figure()` =  Fenster in der grafischen Oberflche (GUI)
- innerhalb dieser figure knnen mehrere *subplots* enthalten sein
    - whrend `subplot()` die plots in regulren Gittern anordnet, erlaubt `axes` ($\neq$ axis **!!**) eine freie Platzierung innerhalb der *figure*


- Beispiel oben: figure und axes/subplots wurden implizit erzeugt
    - wir rufen `plt.plot()` auf
    - matplotlib ruft `gca()` auf ("get current axes")
    - `gca()` ruft `gcf()` auf ("get current figure")
        - falls `None`: `gcf()` ruft `figure()` auf um eine zu erzeugen mit `subplot(1,1,1)`

- das kann man alles explizit steuern





### figures
- Fenster in der grafischen Oberflche mit Titel "Figure #"
- mehrere (optionale) Parameter, zB. die Nummer `num` (#)

````python
plt.figure(num = 3)
plt.close()
````

### subplots
- innerhalb des Fensters knnen (sub)plots in regulrem Gitter angeordnet werden

````python
plt.figure()
plt.subplot(2, 2, 1)
plt.text(0.5, 0.5, 'subplot(2,2,1)', ha='center', va='center', size=24, alpha=.9)

plt.subplot(2, 2, 2)
plt.text(0.5, 0.5, 'subplot(2,2,2)', ha='center', va='center', size=24, alpha=.5)

plt.subplot(2, 2, 3)
plt.text(0.5, 0.5, 'subplot(2,2,3)', ha='center', va='center', size=24, alpha=.5)

plt.subplot(2, 2, 4)
X = np.linspace(0,1,50)
print("X", X)
plt.plot(X, np.sin(X*np.pi*2))

plt.tight_layout()
plt.show()
````

### axes
- *axes* hnlich wie *subplots*, aber knnen berall in der *figure* platziert werden
- Beispielanwendung: einen kleineren plot innerhalb eines greren plots 

````python
plt.figure()
plt.axes([0., 0., .8, .8])
plt.plot(X, np.cos(X*np.pi*2), '-')
plt.text(0.8, .9, 'cosinus', ha='center', va='center', size=30, alpha=.5)

plt.axes([.1, .1, .3, .3])
plt.plot(X, np.sin(X*np.pi))
plt.text(0.5, 0.1, 'sinus', ha='center', va='center', size=16, alpha=.5)

plt.show()
````

### ticks 
 - engl., "Zeichen, Vermerkzeichen, Kontrollzeichen"
 - *tick locators*: spezifizieren wo ticks platziert werden
 - *tick formatters*: Ausssehen formatieren 

 ##### tick locators
 
 Docs: https://matplotlib.org/3.1.1/gallery/ticks_and_spines/tick-locators.html

````python
plt.plot(np.linspace(-1,1,40), np.linspace(-1,1,40)**2)
locators = [    'plt.NullLocator()',
                'plt.MultipleLocator(1.0)',
                'plt.FixedLocator([0, 0.5, 1])',
                'plt.IndexLocator(3, 1)',
                'plt.LinearLocator(5)',
                'plt.LogLocator(2, [1.0])',
                'plt.AutoLocator()',
            ]
locator = locators[-1]
ax = plt.gca()
ax.xaxis.set_major_locator(eval(locator))
````

### Contour plots

````python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.8, cmap='hot')
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
````

### Imshow

- fr Bilder (siehe oben)
- zum visualisieren von $\mathbb{R}^2 \to \mathbb{R}$

````python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 5
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)

plt.imshow(f(X, Y), interpolation = 'Gaussian') # Gaussian
f(X,Y).shape
A = np.eye(10, k=1)
plt.imshow(A, interpolation = 'nearest')
````

### 3d plots

Zum visualisieren von z.B. $\mathbb{R}^2 \to \mathbb{R}$
```python
from mpl_toolkits.mplot3d import Axes3D
```

````python
from mpl_toolkits.mplot3d import Axes3D

figurename = plt.figure()
ax = Axes3D(figurename)
x, y = np.arange(-4, 4, 0.25), np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(x, y)
ax.plot_surface(X, Y, np.sin(np.sqrt(X**2 + Y**2)), cmap='jet') 

ax.contourf(X, Y, np.sin(np.sqrt(X**2 + Y**2)), zdir='z',offset=-2,  cmap='hot')
ax.set_zlim(-2,  2)
plt.show()
````
